```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

tor = pd.read_csv("1950-2022_actual_tornadoes.csv")

tor
```

Figure 1

```python
tor.st.value_counts(ascending=True)[43:].plot(kind='barh',zorder=5)
plt.title("Top 10 US States by Tornado Count (1950-2020)")
plt.xlabel("Number of Tornadoes")
plt.ylabel("State")
plt.grid()
```

```python
tor2000_2020 = tor[tor.yr >= 2000]
tor2000_2020.st.value_counts(ascending=True)[43:].plot(kind='barh',zorder=5)
plt.title("Top 10 US States by Tornado Count (2000-2020)")
plt.xlabel("Number of Tornadoes")
plt.ylabel("State")
plt.grid()
```
